```python
#removing columns that I do not need
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis = 1)
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis = 1)

#checking that the column were dropped
print(dete_survey_updated.columns)
print(tafe_survey_updated.columns)
```

## Creating a New Column

Since my end goal is to answer the question below, I will need a column containing the length of time an employee spent in their workplace, or years of service, in both data frames. 
- End goal: Are employees who have only worked for the institutes for a short period resigning due to some kind of dissatisfaction? What about employees who have been at the job longer?

The `tafe_resignations` data frame already contains a "service" column, which we renamed to institute_service.

Below, I will calculate the years of service in the `dete_survey_updated` data frame by subtracting the `dete_start_date` from the `cease_date` and create a new column named `institute_service`. 


```python
#calculate the lenght of time an employee spent in their respective workplace
#and create a new column
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']

#checking the result
dete_resignations['institute_service'].head()
```

## Identify Dissatisfied Employees

Next, I will indentify any employees who resigned beacuse they were dissatisfied. Below are the columns I will use to categorize employees as dissatisfied from each dataframe:
1. tafe_survey_updated:
    - `Contributing Factors. Dissatisfaction`
    - `Contributing Factors. Job Dissatisfaction`
2. dete_survey_updated
    - `job_dissatisfaction`
    - `dissatisfaction_with_the_department`
    - `physical_work_environment`
    - `lack_of_recognition`
    - `lack_of_job_security`
    - `work_location
    - `employment_conditions`
    - `work_life_balance`
    - `workload`

Suppose the employee indicated any of the factors above causing them to resign. In that case, I will mark them as `dissatisfied` in a new column. After the changes, the new `dissatisfied` column will contain the following:
   - `True`: indicates a person resigned because they were dissatisfied in some way
   - `False`: indicates a person resigned because of a reason other than dissatisfaction with the job
   - `NaN`: indicates the value is missing

```python
#checking the unique values of tafe_resignations
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
#updating the values in the contributing factor columns to be either true, false, or NaN
def update_vals(x):
    if x == '-':
        return False
    elif pd.isnull(x):
        return np.nan
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()

# Check the unique values after the updates
tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
#updating the values in the columns related to dissatisfiaction to be either True, False, or NaN
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

## Combining the Data

Below, I will add an institute column to differentiate the data from each survey after combining them. Then, I will combine the data frames and drop any remaining columns that I do not need.

```python
#adding the institute column
dete_resignations_up['institutes'] = 'DETE'
tafe_resignations_up['institutes'] = 'TAFE'
```

```python
#combining the data frames
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index = True)

#verifying the number of non null values in each column
combined.notnull().sum().sort_values()
```

```python
#drop columns with less than 500 non null values
combined_updated = combined.dropna(thresh = 500, axis = 1).copy()
```

## Cleaning the Service Column

Next, I will clean the `institute_service` column and categorize employees according to the following definitions:

- New: Less than 3 years in the workplace
- Experienced: 3-6 years in the workplace
- Established: 7-10 years in the workplace
- Veteran: 11 or more years in the workplace

The analysis is based on [this article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage), which makes the argument that understanding employee's needs according to career stage instead of age is more effective. 

```python
#checking the unique values of combined data for the categories of work experience
combined_updated['institute_service'].value_counts(dropna = False)
```

```python
#extracting the years of service and converting the type to float
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')

#checking to see whether the years are correctly extracted
combined_updated['institute_service_up'].value_counts()
```

```python
#converting the years of service to categories
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)

#checking the update
combined_updated['service_cat'].value_counts()
```

## Performing Initial Analysis

Lastly, I will replace the missing values in the `dissatisfied` column with the `False` value, the most frequent value. Then, I will calculate the percentage of employees who resigned due to dissatisfaction in each `service_cat` group and plot the results. 

Note that since I still have additional missing values left to deal with, this will be an initial introduction to the analysis, not the final analysis. 

```python
#verifying unique values
combined_updated['dissatisfied'].value_counts(dropna = False)
```

```python
#replacing missing values with the most frequent value
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

```python
#calculating the percentage of employees who resigned due to dissatisfaction in each category
dis_pct = combined_updated.pivot_table(index='service_cat', values='dissatisfied')

#plotting the results
%matplotlib inline
dis_pct.plot(kind='bar', rot=30)
```

## Conclusion

From the initial analysis above, I can tentatively conclude that employees with seven or more years of service are more likely to resign due to some kind of dissatisfaction with the job than employees with less than seven years of service. ## Cleaning the Age column and looking for differences

Below, I will follow exactly what I did for the service column to clean the age column. 

```python
combined_updated['age'].value_counts(dropna = False)
```

```python
#extracting the age and converting the type to float
combined_updated['age_up'] = combined_updated['age'].astype('str').str.extract(r'(\d+)')
combined_updated['age_up'] = combined_updated['age_up'].astype('float')

#checking to see whether the ages are correctly extracted
combined_updated['age_up'].value_counts()
```

```python
#converting the age to categories
def transform_age(val):
    if val >= 51:
        return "Experienced Adult"
    elif 30 <= val < 51:
        return "Middle Aged Adult"
    elif 20 <= val < 30:
        return "Young Adult"
    elif pd.isnull(val):
        return np.nan
    else:
        return "Barely an Adult"
combined_updated['age_cat'] = combined_updated['age_up'].apply(transform_age)

#checking the update
combined_updated['age_cat'].value_counts()
```

```python
#calculating the percentage of employees who resigned due to dissatisfaction in each category
dis_pct = combined_updated.pivot_table(index='age_cat', values='dissatisfied')

#plotting the results
%matplotlib inline
dis_pct.plot(kind='bar', rot=30)
```

## Conclusion Two

From the age analysis, I can tentatively conclude that more 51+ years of age group resign more because they are dissatisfied than the other groups. Additionally, it appears that middle-aged adults (30-50) years leave because of dissatisfaction more than young adults. However, the difference seems to be small. I could run a statistical analysis to see if these differences are statistically significant.
